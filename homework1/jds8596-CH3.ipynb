```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
def MonteCarlo(p):
    counter = 0
    for i in range(p):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if (x*x + y*y < 1):
            counter = counter + 1
    return 4*counter/p
plist = []
trueerrorlist = []
relativeerrorlist = []
for p in range(10, 10001, 20):
    calc_value = MonteCarlo(p)
    trueerror = abs(np.pi-calc_value) 
    plist.append(p)
    trueerrorlist.append(trueerror)
    relativeerror = abs((np.pi-calc_value)/np.pi)
    relativeerrorlist.append(relativeerror)
    
plt.plot(plist, trueerrorlist)
plt.title('Absolute Error')
plt.show()

plt.plot(plist, relativeerrorlist)
plt.title('True Error')
plt.show()
```
