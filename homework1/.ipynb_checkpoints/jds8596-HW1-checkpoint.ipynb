```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
def MonteCarlo(p):
    counter = 0
    for i in range(p):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if (x*x + y*y < 1):
            counter = counter + 1
    return 4*counter/p
plist = []
errorlist = []
for p in range(10, 10001, 20):
    calc_value = MonteCarlo(p)
    error = abs(calc_value - np.pi) 
    plist.append(p)
    errorlist.append(error)
    
plt.plot(plist, errorlist)
plt.show()
```
