```python
import math
def calc_hyp(a, b):
    h = math.sqrt(a*a + b*b)
    return h
```

```python
my_h = calc_hyp(3, 4)
print(my_h)
```

```python
import random
import math
def MonteCarlo(shotstaken):
    counter = 0
    for i in range(0, shotstaken):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if (x*x + y*y < 1):
            counter = counter + 1
    return 4*counter/shotstaken
        
   
    
```

```python
MonteCarlo(50)
```

```python
MonteCarlo(1000000000)
```

```python
MonteCarlo(10000)
```
