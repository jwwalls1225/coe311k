```python
import random
import math
import numpy as np
def MonteCarlo(shotstaken):
    counter = 0
    for i in range(0, shotstaken):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if (x*x + y*y < 1):
            counter = counter + 1
    return 4*counter/shotstaken
for i in range(0, 100):
    calc_pi = MonteCarlo(1000)
    true_pi = np.pi
    true_error = abs(true_pi - calc_pi)
    relative_error = abs((true_pi - calc_pi)/(true_pi))
    print(true_error, relative_error)
```
